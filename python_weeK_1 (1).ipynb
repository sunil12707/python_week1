#  **comparison operators**

Equal to (==): Checks if the values of two operands are equal.
Example: x == y (Returns True if x is equal to y)

Not equal to (!=): Checks if the values of two operands are not equal.
Example: x != y (Returns True if x is not equal to y)

Greater than (>): Checks if the value of the left operand is greater than the value of the right operand.
Example: x > y (Returns True if x is greater than y)

Less than (<): Checks if the value of the left operand is less than the value of the right operand.
Example: x < y (Returns True if x is less than y)

Greater than or equal to (>=): Checks if the value of the left operand is greater than or equal to the value of the right operand.
Example: x >= y (Returns True if x is greater than or equal to y)

Less than or equal to (<=): Checks if the value of the left operand is less than or equal to the value of the right operand.
Example: x <= y (Returns True if x is less than or equal to y)

```python
# Comparison operators example
x = 10
y = 5

# Equal to
print(x == y)  # False

# Not equal to
print(x != y)  # True

# Greater than
print(x > y)  # True

# Less than
print(x < y)  # False

# Greater than or equal to
print(x >= y)  # True

# Less than or equal to
print(x <= y)  # False
```

# **Bitwise opertor**

```python
# Bitwise AND (&): Performs a bitwise AND operation on two operands.

# Bitwise OR (|): Performs a bitwise OR operation on two operands.

# Bitwise XOR (^): Performs a bitwise exclusive OR operation on two operands.

# Bitwise NOT (~): Performs a bitwise NOT operation on a single operand.

# Left Shift (<<): Shifts the bits of the left operand to the left by the number of positions specified by the right operand. Zeros are shifted in from the right side.

# Right Shift (>>): Shifts the bits of the left operand to the right by the number of positions specified by the right operand. The sign bit is shifted in from the left side for signed integers, and zeros are shifted in for unsigned integers.
```

```python
bin(10)
```

```python
bin(7)
```

```python
# Bitwise AND (&)
a = 10  # 1010 in binary
b = 7   # 0111 in binary
result_and = a & b
print(f"Bitwise AND: {result_and}")  # Output: 2 (0010 in binary)

# Bitwise OR (|)
a = 10  # 1010 in binary
b = 7   # 0111 in binary
result_or = a | b
print(f"Bitwise OR: {result_or}")  # Output: 15 (1111 in binary)

# Bitwise XOR (^)
a = 10  # 1010 in binary
b = 7   # 0111 in binary
result_xor = a ^ b
print(f"Bitwise XOR: {result_xor}")  # Output: 13 (1101 in binary)

# Bitwise NOT (~)
a = 10  # 1010 in binary
result_not = ~a
print(f"Bitwise NOT: {result_not}")  # Output: -11

# Left Shift (<<)
a = 5   # 101 in binary
shift_amount = 2
result_left_shift = a << shift_amount
print(f"Left Shift: {result_left_shift}")  # Output: 20 (10100 in binary)
 
# Right Shift (>>)
a = 20  # 10100 in binary
shift_amount = 2
result_right_shift = a >> shift_amount
print(f"Right Shift: {result_right_shift}")  # Output: 5 (101 in binary)
```

```python
# Bitwise NOT (~)
a = 10  # 1010 in binary
result_not = ~a
print(f"Bitwise NOT: {result_not}")  # Output: -11
```
